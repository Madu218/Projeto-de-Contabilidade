```python
import pandas

dataFrame = pandas.read_excel("real-bd.xlsx", skiprows = range(0, 4)) # Base de dados
dF = dataFrame[["Instituição", "UF", "Conta", "Valor"]] # Base de dados com as colunas que iremos utilizar

# Áreas que iremos somar as despesas
variaveis = (dF["Conta"].str.startswith("10")) | (dF["Conta"] == "06 - Segurança Pública") | (dF["Conta"] == "08 - Assistência Social") | (dF["Conta"] == "09 - Previdência Social") | (dF["Conta"] == "12 - Educação")

despesas_selecionadas = dF.loc[variaveis] # Base de dados com as áreas filtradas
despesas_selecionadas["Valor"] = pandas.to_numeric(despesas_selecionadas["Valor"], errors="coerce") # Transforma em NaN os valores que não podem ser somados por estarem vazios


despesas_por_estado = despesas_selecionadas[["UF", "Conta", "Valor"]].groupby(["UF", "Conta"]).sum()
despesas_por_municipio = despesas_selecionadas[["Instituição", "UF", "Conta", "Valor"]].groupby(["UF", "Instituição", "Conta"]).sum()

dF['Valor'].fillna(0, inplace=True) # Transforma todos os valores NaN da coluna em 0

display(despesas_por_estado)
display(despesas_por_municipio)

despesas_selecionadas.to_excel('TabelaFinal.xlsx', index=False)
```

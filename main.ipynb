```python
import pandas

dataFrame = pandas.read_excel("bd.xlsx", skiprows = range(0, 4)) # Base de dados
dF = dataFrame[["UF", "Conta", "Valor"]] # Base de dados com as colunas que iremos utilizar

# Áreas que iremos somar as despesas
variaveis = (dF["Conta"].str.startswith("10")) | (dF["Conta"] == "06 - Segurança Pública") | (dF["Conta"] == "08 - Assistência Social") | (dF["Conta"] == "09 - Previdência Social") | (dF["Conta"] == "12 - Educação")

despesas_selecionadas = dF.loc[variaveis] # base de dados com as áreas filtradas

despesas_por_estado = despesas_selecionadas[["UF", "Conta", "Valor"]].groupby(["UF", "Conta"]).sum() # soma as despesas desejadas por estado

display(despesas_por_estado)
```
